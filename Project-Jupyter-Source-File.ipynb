```python
# Select only numeric columns for IQR calculation
numeric_columns = df.select_dtypes(include=["number"])

# Calculate the Q1 (25th percentile) and Q3 (75th percentile)
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1
print("Interquartile Range (IQR):\n", IQR)

# Filter out outliers using the IQR
df = df[~((numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))).any(axis=1)]

# Check the new shape of the dataframe
print("Shape after removing outliers:", df.shape)
```

# *1. Pie Chart*
*Explanation:* A pie chart is an effective way to represent proportions within categorical data. In this visualization, the chart depicts the distribution of various drive modes (Drivetrain) across the dataset. Each segment of the pie corresponds to a specific drivetrain type, with the size of the segment proportional to its occurrence in the dataset. The percentage labels and vibrant colors (#ff9999, #66b3ff, #99ff99, #ffcc99) enhance readability, making it simple to identify the dominant drive modes. The clear segmentation provides insights into which drivetrain types are most popular, enabling quick and intuitive comparisons among categories.

```python
df['Drive Mode'].value_counts().plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Distribution of Drive Modes")
plt.ylabel('')
plt.tight_layout()
#Pie Chart
```

# *2. Line Plot*
*Explanation:* A line plot is a great choice for visualizing trends and patterns over time. In this case, the chart represents the average price of cars (Price) over the years (Year). Each point on the line corresponds to the mean price for a given year, connected by a smooth green line with circular markers (o) for better readability. The grid enhances interpretability by allowing for precise tracking of values along both axes. This visualization effectively highlights fluctuations or steady trends in car prices, making it easier to observe historical pricing behaviors and identify potential significant shifts in the automotive market.

```python
df.groupby('Year')['Price'].mean().plot(kind='line', figsize=(10, 5), color='green', marker='o')
plt.title("Average Price of Cars Over the Years")
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.tight_layout()
#Line Plot
```

# *3. Horizontal Bar Chart*
*Explanation:* The horizontal bar chart is a useful visualization technique for highlighting categorical data in a comparative manner. In this case, the chart displays the top 20 car manufacturers based on the number of cars they produce. Each bar represents a car manufacturer, and its length corresponds to the total number of cars associated with that manufacturer. This visualization is effective in clearly identifying manufacturers with the highest counts, making it easier to spot dominant players in the automobile market. With this approach, it becomes visually intuitive to determine which manufacturers lead in terms of production.

```python
df.Make.value_counts().nlargest(20).plot(kind='barh', figsize=(10, 5), color='skyblue')
plt.title("Number of Cars by Make (Horizontal)")
plt.xlabel('Number of Cars')
plt.ylabel('Make')
plt.tight_layout()
#Horizontal Bar Chart
```

# *4. Histogram*
*Explanation:* # A histogram provides an insightful way to understand the distribution of numerical data. In this case, the chart illustrates the frequency of car prices (Price) across intervals. The data is divided into 30 bins, which group prices into ranges, and each bar represents the number of cars falling within that specific price range. The orange bars are outlined with black edges for added clarity, ensuring the intervals are well-defined and visually distinct. This visualization helps to identify key patterns, such as the most common price range or the spread of prices in the dataset, providing a deeper understanding of the pricing landscape in the automotive industry.

```python
df['Price'].plot(kind='hist', bins=30, figsize=(10, 5), color='orange', edgecolor='black')
plt.title("Distribution of Car Prices")
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
#Histogram
```

# *5. Box Plot*
*Explanation:* A box plot is an excellent tool for visualizing the spread and variability of a numerical feature across different categories. In this case, the chart shows how car prices (Price) are distributed across various Vehicle Size categories. Each box represents a category and contains important statistical information, such as the median (central line), the interquartile range (box edges), and potential outliers (dots outside the whiskers). The patch_artist parameter adds visual appeal with filled colors, making it easier to differentiate between categories. This plot is particularly useful for identifying patterns, such as which vehicle sizes tend to have higher prices or which exhibit greater variability, enabling meaningful insights into the relationship between vehicle size and pricing.

```python
df.boxplot(column='Price', by='Vehicle Size', figsize=(10, 5), patch_artist=True)
plt.title("Price Distribution by Vehicle Size")
plt.suptitle("")  # Suppress the default title
plt.xlabel('Vehicle Size')
plt.ylabel('Price')
plt.tight_layout()
#Box Plot
```

# *6. Stacked Bar Chart*
*Explanation:* The stacked bar chart is a powerful tool for visualizing the distribution of categorical variables within groups. In this example, the chart showcases the count of different Transmission types for each Vehicle Size. Each bar represents a vehicle size category, and its segmentss illustrate the proportion of each transmission type (Transmission) within that category. The stacked design allows for easy comparison both within and across categories, providing insights into how transmission types vary by vehicle size. The use of the viridis colormap enhances the visual appeal by creating a smooth and distinct color gradient for each transmission type. This chart effectively highlights patterns, such as the dominance of certain transmission types in specific vehicle sizes, enabling a deeper understanding of the relationship between transmission and vehicle size characteristics.

```python
df.groupby(['Vehicle Size', 'Transmission']).size().unstack().plot(kind='bar', stacked=True, figsize=(10, 5), colormap='viridis')
plt.title("Transmission Types Across Vehicle Sizes")
plt.xlabel('Vehicle Size')
plt.ylabel('Count')
plt.tight_layout()
#Stacked Bar Chart
```

# *7. Kernel Density Estimation (KDE) plot*
*Explanation:* # A Kernel Density Estimation (KDE) plot is a powerful tool for visualizing the probability density of a continuous variable. In this example, the plot illustrates the density distribution of highway miles per gallon (MPG_Highway). The smooth red curve provides an estimate of the data's underlying distribution, making it easy to identify common values and the spread of MPG_Highway. This visualization is particularly useful for understanding the central tendency, skewness, and variance within the dataset. The continuous nature of the curve allows for a clearer representation compared to a histogram, offering insights into the efficiency of vehicles on highways. The streamlined layout ensures readability and helps interpret the distribution effectively.

```python
df['MPG-H'].plot(kind='kde', figsize=(10, 5), color='red')
plt.title("Density Plot of Highway MPG")
plt.xlabel('Highway MPG')
plt.ylabel('Density')
plt.tight_layout()
#KDE Plot
```
